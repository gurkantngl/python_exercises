```python
# survived, pclass ve cinsiyet değişkenlerine göre kırılımında sum, count ve mean değerlerini bulun
result = df.groupby(['survived', 'pclass', 'sex']).agg({
    'survived': 'sum',
    'pclass': 'count',
}).rename(columns={'survived': 'Toplam Hayatta Kalan', 'pclass': 'Toplam Yolcu Sayısı'})

result['Ortalama Hayatta Kalma Oranı'] = result['Toplam Hayatta Kalan'] / result['Toplam Yolcu Sayısı'] * 100

print(result)
```

```python
data = sns.load_dataset("Tips")
```

```python
data.head()
```

```python
result = data.groupby('time')['total_bill'].agg(['sum', 'min', 'max', 'mean'])
```

```python
result
```

```python
result = data.groupby(['day', 'time'])['total_bill'].agg(['sum', 'min', 'max', 'mean'])
```

```python
result
```

```python
# Lunch zamanına ve kadın müşterilere ait total_bill ve tip değerlerinin day'e göre toplamını, min, max ve ortalamasını bulun
lunch_female_data = data[(data['time'] == 'Lunch') & (data['sex'] == 'Female')]

result = lunch_female_data.groupby('day')[['total_bill', 'tip']].agg(['sum', 'min', 'max', 'mean'])
```

```python
result
```

```python
# Filtrelenmiş verilerin "total_bill" sütununun ortalamasını bulun
mean_total_bill = data.loc[(data['size'] < 3) & (data['total_bill'] > 10)]['total_bill'].mean()
```

```python
print(f"Ortalama total bill: {mean_total_bill}")
```

```python
data['total_bill_tip_sum'] = data['total_bill'] + data['tip']

# Sonuçları kontrol edin
print(data[['total_bill', 'tip', 'total_bill_tip_sum']].head())
```
